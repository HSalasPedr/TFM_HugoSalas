### Parsing genes
Parsing genes from models coming from DSM319 and _B. subtilis_ str 168 to see wether there are genes to add to our model coming from _B. subtilis_ str 168, due to de homolgy threshold

```python
import pandas as pd
```

```python
bsubt = pd.read_csv('Bsubt2_vs_PmegMOA_parsed.csv')
dsm = pd.read_csv('DSM319_vs_PmegMOA_parsed.csv')
```

```python
PIDbsubt = bsubt[['gene', 'subject', 'PID']]
PIDdsm = dsm[['gene', 'subject', 'PID']]
```

```python
df = pd.merge(PIDbsubt, PIDdsm, on = ['subject'])
```

```python
df
```

```python
out = df[(df['PID_x'] > 60) & (df['PID_y'] < 80)]
```

```python
out2 = df[(df['PID_x'] > 60) & (df['PID_y'] < 80) & (df['PID_y'] > 60)]
```

```python
out3 = df[(df['PID_x'] > 60) & (df['PID_x'] < 80) & (df['PID_y'] < 80) & (df['PID_y'] > 60)]
```

```python
out3
```

```python
out.to_csv('PmegMOA_fromBsubt_GenesToAdd.csv')
```

```python
out2.to_csv('PmegMOA_fromBsubt_GenesToAdd2.csv')
```

```python
out3.to_csv('PmegMOA_fromBsubt_GenesToAdd3.csv')
```

Now we check how many of the genes in out1 and out2 are not in our PmegMOA model coming from _B. subtilis_ str 168

```python
import cobra
from cobra.io import read_sbml_model
```

```python
model = read_sbml_model('PmegMOA_fromBsubt.xml')
model
```

```python
genelist_model = set([model.genes[i]._id for i in range(1, len(model.genes))])
genelist_model
```

```python
genelist_add_1 = set(out['gene_x'].to_numpy())
```

```python
genes_to_add_1 = genelist_add_1.difference(genelist_model)
print(len(genes_to_add_1))
```

```python
genelist_add_2 = set(out2['gene_x'].to_numpy())
genes_to_add_2 = sorted(genelist_add_2.difference(genelist_model))
print(len(genes_to_add_2))
```

```python
all_genes_to_add = genes_to_add_1.union(genes_to_add_2)
print(len(all_genes_to_add))
```

```python
df_out = pd.DataFrame({'genes_to_add': list(all_genes_to_add)})
df_out
```

```python
df_out.to_csv('genes_to_add.csv')
```

```python
df_out2 = df.loc[df['gene_x'].isin(df_out['genes_to_add'])]
```

```python
df_out2 = df_out2[(df_out2['PID_x'] > 60) & (df_out2['PID_y'] < 80) & (df_out2['PID_y'] > 60)]['gene_y']
df_out2
```

```python
df_out2.to_csv('Genes_to_add_dsm.csv')
```
