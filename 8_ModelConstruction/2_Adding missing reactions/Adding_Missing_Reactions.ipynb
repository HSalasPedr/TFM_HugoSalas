```python
rxn_dict = dict()

for row in rxns_to_add.iterrows():
    #rxn_dict[row[1]['gene_y']] = set()
    rxn_dict[row[1]['genes_to_add']] = set()
```

```python
def delete_non_alphanumeric(string):
    pattern = r'[^a-zA-Z0-9_]'  # Matches any character that is not a letter, digit, or underscore
    cleaned_string = re.sub(pattern, '', string)
    return cleaned_string
```

```python
#for row in modDSM.iterrows():
for row in modBsubt.iterrows():
    #gene_list, index = row[1]['GPR'], modDSM[modDSM['Abbreviation'] == row[1]['Abbreviation']].index[0]
    gene_list, abb = row[1]['GPR'], row[1]['Abbreviation']
        
    if not pd.isna(gene_list):        
        gene_list = gene_list.split(' ')
        
        aux_gene_list = []
        
        for word in gene_list:        
            #if 'BMD_' in word:
            if 'BSU_' in word:
                aux_gene_list.append(word)
                
        fixed_gene_list = list()
        
        for word in aux_gene_list:
            fixed_gene_list.append(delete_non_alphanumeric(word))
                
        #gene_list = aux_gene_list
        gene_list = fixed_gene_list
        
        for gene in gene_list:
            try:
                #rxn_dict[gene].add(index)
                rxn_dict[gene].add(abb)  
            except:
                pass
```

```python
rxn_dict
```

```python
genes_to_add_set = set()

for key in rxn_dict.keys():
    genes_to_add_set.update(rxn_dict[key])

genes_to_add_set
```

```python
final_rxns_to_add = modDSM.loc[modDSM['Abbreviation'].isin(list(genes_to_add_set))]
```
